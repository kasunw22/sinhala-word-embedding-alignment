```python
def get_lang_code(lang_name):
  if lang_name in LANGUAGE_NAMES:
    return lang_name

  lang_name_lower = lang_name.lower()

  for code, name_list in LANGUAGE_NAMES.items():
    if (lang_name in name_list) or (lang_name_lower in name_list):
      print(f"{lang_name} --> {code}")
      return code
  else:
    raise f"Language {lang_name} could not be found!"
```
