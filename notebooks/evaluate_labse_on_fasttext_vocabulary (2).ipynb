```python
def load_vocab(file_path, maxload=None):
    with open(file_path) as f:
        vocab = [x.strip() for x in f.readlines()]

    n_vocab = len(vocab)

    if maxload is None:
        maxload = n_vocab

    print(f"[INFO] Found {len(vocab)} entries and, returning {min(maxload, n_vocab)} entries...")

    return vocab[:maxload]
```

```python
def compute_embeddings(sentences):
    t0 = time.time()
    embeddings = model.encode(sentences, convert_to_tensor=True)
    print(f"[INFO] Encoding time {(time.time() - t0) * 1000} ms")

    print(embeddings.shape, embeddings[0].shape)

    return embeddings
```

```python
def wi(words):
    w2i = {}
    i2w = {}
    for i, w in enumerate(words):
        if w not in w2i:
            w2i[w] = i
            i2w[i] = w

    return w2i, i2w
```

```python
def idx(words):
    w2i = {}
    for i, w in enumerate(words):
        if w not in w2i:
            w2i[w] = i

    return w2i


def load_lexicon(filename, words_src, words_tgt, verbose=True):
    f = io.open(filename, 'r', encoding='utf-8')
    lexicon = collections.defaultdict(set)
    idx_src , idx_tgt = idx(words_src), idx(words_tgt)
    vocab = set()

    for line in f:
        # print(line)
        l_split = line.split()
        word_src, word_tgt = l_split[0], " ".join(l_split[1:])
        if word_src in idx_src and word_tgt in idx_tgt:
            lexicon[idx_src[word_src]].add(idx_tgt[word_tgt])
            # print(f"[DEBUG] added {word_src}, {word_tgt}")
        vocab.add(word_src)

    if verbose:
        coverage = len(lexicon) / float(len(vocab))
        print("[INFO] Coverage of source vocab: %.4f" % (coverage))

    return lexicon, float(len(vocab))
```

```python
def lematized_match(ref, pred):
    # ref = lexicon[idx_src[j]]
    # pred = pred[j - i]

    if not isinstance(ref, list):
        ref = list(ref)

    if not isinstance(pred, list):
        pred = list(pred)

    # any(one_top_k in list(lexicon[idx_src[j]]) for one_top_k in pred[j - i])

    split_ref = []

    for x in ref:

      if isinstance(x, torch.Tensor):
        x = x.item()

      if x in si_splits:
        split_ref.append(si_splits[x])
      else:
        try:
          split = word_splitter.split(words_tgt[x])['base']
        except:
          continue

        try:
          split = idx_tgt[split]
        except:
          ...

        split_ref.append(split)
        si_splits[x] = split

        with open(si_splits_file_path, 'wb') as f:
            pickle.dump(si_splits, f)

    split_pred = []

    for y in pred:

      x = y.item()

      if x in si_splits:
        split_pred.append(si_splits[x])
      else:
        try:
          split = word_splitter.split(words_tgt[x])['base']
        except:
          continue

        try:
          split = idx_tgt[split]
        except:
          ...

        split_pred.append(split)
        si_splits[x] = split

        with open(si_splits_file_path, 'wb') as f:
            pickle.dump(si_splits, f)

    split_ref = ref + split_ref
    split_pred = pred + split_pred

    return any(one_top_k in split_ref for one_top_k in split_pred)
```

```python
def compute_nn_accuracy(x_src, x_tgt, lexicon, bsz=100, lexicon_size=-1, top_k=1, norm=False):  # use norm=False if vectors are already normalized
    if lexicon_size < 0:
        lexicon_size = len(lexicon)

    idx_src = list(lexicon.keys())
    acc = 0.0

    if norm:
        # x_src /= np.linalg.norm(x_src, axis=1)[:, np.newaxis] + 1e-8
        for i in range(len(x_src)):
            x_src[i] /= np.linalg.norm(x_src[i]) + 1e-8

    if norm:
        # x_tgt /= np.linalg.norm(x_tgt, axis=1)[:, np.newaxis] + 1e-8
        for i in range(len(x_tgt)):
            x_tgt[i] /= np.linalg.norm(x_tgt[i]) + 1e-8

    for i in range(0, len(idx_src), bsz):
        e = min(i + bsz, len(idx_src))
        scores = cos_sim(x_tgt, x_src[idx_src[i:e]])
        # print(scores.shape)

        # pred = scores.argsort(dim=0, descending=False)[-top_k:].T
        pred = scores.argsort(dim=0, descending=True)[:top_k].T
        # print(pred.shape)
        for k, j in enumerate(range(i, e)):
            # print(f"Pair: {words_src[idx_src[j]]} === {[words_tgt[tg] for tg in lexicon[idx_src[j]]]} ({lexicon[idx_src[j]]}) =================")

            if any(one_top_k in list(lexicon[idx_src[j]]) for one_top_k in pred[j - i]):
                acc += 1.0

            elif tgt == "si" and do_lematized_match and lematized_match(ref=list(lexicon[idx_src[j]]), pred=pred[j - i]):
                acc += 1.0

            # for one_top_k in pred[j - i]:
            #     print(f"{words_src[idx_src[j]]} -- {tgt_vocab[one_top_k]} ({one_top_k}) --> {scores.T[k][one_top_k]}")
            #     if one_top_k in list(lexicon[idx_src[j]]):
            #         acc += 1.0
            #         print(f"{one_top_k}*******************************************")
            #         break
        print(f"{i + bsz} done -- correct: {acc}")
        # if i > 0:
        #     break
    print(f"correct: {acc}")
    return acc / lexicon_size
```

```python
def compute_csls_accuracy(x_src, x_tgt, lexicon, lexicon_size=-1, k=10, bsz=1024, top_k=1, norm=False): #, norm=False, norm_src_only=False):
    if lexicon_size < 0:
        lexicon_size = len(lexicon)

    idx_src = list(lexicon.keys())

    if norm:
      # x_src /= np.linalg.norm(x_src, axis=1)[:, np.newaxis] + 1e-8
      for i in range(len(x_src)):
          x_src[i] /= np.linalg.norm(x_src[i]) + 1e-8

    if norm:
      # x_tgt /= np.linalg.norm(x_tgt, axis=1)[:, np.newaxis] + 1e-8
      for i in range(len(x_tgt)):
          x_tgt[i] /= np.linalg.norm(x_tgt[i]) + 1e-8

    sr = x_src[list(idx_src)]
    # sc = np.dot(sr, x_tgt.T)
    sc = torch.matmul(sr, x_tgt.T)
    similarities = 2 * sc
    # sc2 = np.zeros(x_tgt.shape[0])
    sc2 = torch.zeros(x_tgt.shape[0]).to(similarities.device)

    for i in range(0, x_tgt.shape[0], bsz):
        j = min(i + bsz, x_tgt.shape[0])
        # sc_batch = np.dot(x_tgt[i:j, :], x_src.T)
        sc_batch = torch.matmul(x_tgt[i:j, :], x_src.T)
        # dotprod = np.partition(sc_batch, -k, axis=1)[:, -k:]
        dotprod = torch.topk(sc_batch, k=k, dim=1).values
        # sc2[i:j] = np.mean(dotprod, axis=1)
        sc2[i:j] = torch.mean(dotprod, dim=1)
        torch.cuda.empty_cache()

        # print(f"{i + bsz} done")
    torch.cuda.empty_cache()
    # similarities -= sc2[np.newaxis, :]
    similarities -= torch.unsqueeze(sc2, dim=0)

    del sc_batch, dotprod, sc2
    torch.cuda.empty_cache()

    nn = []
    # # nn = np.argmax(similarities, axis=1).tolist()
    # # nn = np.argsort(similarities, axis=1).T[-top_k:][::-1].T
    # nn = torch.argsort(similarities, axis=1).T[-top_k:].flipud().T
    for sim_vec in similarities:
      nn.append(torch.argsort(sim_vec, dim=-1)[-top_k:].flipud())
    # print(nn)
    correct = 0.0

    for k in range(0, len(lexicon)):

        # if nn[k] in lexicon[idx_src[k]]:
        if any(one_top_k.item() in lexicon[idx_src[k]] for one_top_k in nn[k]):
            correct += 1.0

        elif tgt == "si" and do_lematized_match and lematized_match(ref=lexicon[idx_src[k]], pred=nn[k]):
            correct += 1.0

    return correct / lexicon_size
```

```python
def find_top_k_for_src_word(test_word, target_word, x_src, x_tgt, idx_src, words_src, idx_tgt, words_tgt, top_k=5):
    if isinstance(test_word, str):
        src_embd = x_src[idx_src[test_word]]
        word_str = test_word
    else:
        src_embd = x_src[test_word]
        word_str = words_src[test_word]

    if isinstance(target_word, str):
        tgt_word_idx = idx_tgt[target_word]
        tgt_word_str = target_word
    else:
        tgt_word_str = words_tgt[target_word]
        tgt_word_idx = target_word

    scores = cos_sim(x_tgt, src_embd)

    print(scores.shape)

    # pred = scores.argsort(dim=0, descending=False)[-top_k:].T
    pred = scores.argsort(dim=0, descending=True)[:top_k].T
    print(pred, pred.shape)

    for one_top_k in pred[0]:
        one_top_k = one_top_k.item()
        print(f"{word_str} -- {words_tgt[one_top_k]} --> {scores.T[0][one_top_k]}")

    print(f"\n{word_str} -- {tgt_word_str} --> {scores.T[0][tgt_word_idx]}================")


```

```python
def not_ASCII(word):
    is_ascii = [ord(x)<256 for x in word]

    if all(is_ascii):
        return False

    return True
```

```python
def load_saved_embeddings(embed_path, prune_ascii=False, vocab=None):

    if os.path.isfile(embed_path):
        print(f"[INFO] Loading the embeddings from {embed_path}...")
        embeds = torch.load(embed_path)
        print("[INFO] Loading done...")
    else:
        print(f"[INFO] {embed_path} does not exist. Calculating the embeddings..")
        embeds = compute_embeddings(vocab)
        torch.save(embeds, embed_path)
        print(f"[INFO] Saved the embeddings at {embed_path}...")

    print(f"[INFO] Original embeddings size: {embeds.shape}")

    if prune_ascii and vocab is not None:
        print("[INFO] Pruning the vocabulary by removing ASCII entries...")
        prune_idx = list(map(not_ASCII, vocab))
        embeds = embeds[prune_idx]
        vocab = list(np.array(vocab)[prune_idx])

        print(f"[INFO] After pruning embeddings size: {embeds.shape}")

    return embeds, vocab
```
